```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
```

```python
import pandas as pd
data=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
print(data.shape)
data.head()
```

```python
data.describe(include='all')
```

```python
data=data.drop(labels=['PassengerId','Name','Ticket','Cabin'],axis=1)
testid=test.PassengerId
test=test.drop(labels=['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
print(type(data.Sex))
print(type(data.Embarked))
```

```python
data['Sex']=data['Sex'].map({'female':1,'male':0}).astype(int)
test['Sex']=test['Sex'].map({'female':1,'male':0}).astype(int)

data['Embarked']=data['Embarked'].fillna('S')
test['Embarked']=test['Embarked'].fillna('S')

data['Embarked']=data['Embarked'].map({'Q':2,'C':1,'S':0}).astype(int)
test['Embarked']=test['Embarked'].map({'Q':2,'C':1,'S':0}).astype(int)

data['Fare']=data['Fare'].fillna(32)
test['Fare']=test['Fare'].fillna(32)

data['Age']=data['Age'].fillna(32)
test['Age']=test['Age'].fillna(32)
```

```python
data.info()
```

```python
train_y=data.Survived
train_X=data.drop('Survived',1)
```

```python
from sklearn.model_selection import train_test_split

X,x,Y,y=train_test_split(train_X,train_y)
y=y.values
```

```python
from sklearn.ensemble.gradient_boosting import GradientBoostingClassifier as gb
model=gb()
model.fit(X,Y)
predicted=model.predict(x)
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y,predicted))
```

```python
test.info()
```

```python
new=model.predict(test)
```

```python
submission=pd.DataFrame({'PassengerId':testid,'Survived':new})
submission.to_csv("titanic_submission.csv",index=False)
```
